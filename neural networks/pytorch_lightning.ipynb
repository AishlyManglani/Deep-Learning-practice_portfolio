```python
# Import necessary libraries
import torch
import torch.nn as nn
import torch.optim as optim
import pytorch_lightning as pl
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
torch.manual_seed(0)
X = torch.rand(100, 3)  # 3 input features
y = torch.sin(X[:, 0]) + torch.cos(X[:, 1]) + torch.tanh(X[:, 2]) + torch.normal(0, 0.1, (100,))

# Define the neural network class
class ThreeLayerNN(pl.LightningModule):
    def __init__(self):
        super(ThreeLayerNN, self).__init__()
        self.hidden1 = nn.Linear(3, 5)
        self.hidden2 = nn.Linear(5, 5)
        self.output = nn.Linear(5, 1)
        self.criterion = nn.MSELoss()

    def forward(self, x):
        x = torch.sigmoid(self.hidden1(x))
        x = torch.sigmoid(self.hidden2(x))
        return torch.sigmoid(self.output(x))

    def training_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = self.criterion(y_hat.squeeze(), y)
        return loss

    def configure_optimizers(self):
        return optim.SGD(self.parameters(), lr=0.01)
```

```python
# Create a DataLoader
from torch.utils.data import DataLoader, TensorDataset
dataset = TensorDataset(X, y)
train_loader = DataLoader(dataset, batch_size=32)

# Initialize the model
model = ThreeLayerNN()

# Train the model
trainer = pl.Trainer(max_epochs=1000)
trainer.fit(model, train_loader)

# Final output
print("Final output:", model(X))
```
